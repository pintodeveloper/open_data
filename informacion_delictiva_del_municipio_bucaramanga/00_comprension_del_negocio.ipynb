```python
# 2.Revisar calidad

display(df.isna().sum().sort_values(ascending=False))
print("Duplicados: ", df.duplicated().sum())
```

```python
# trabajar siempre desde una copia

df_limpio = df.copy()

df_limpio.columns = (
    df_limpio.columns.str.strip()
    .str.lower()
    .str.replace(r"\s+", "_", regex=True)
)
```

```python
#Quitar espacios y uniformar texto

columnas_texto = df_limpio.select_dtypes(include="object").columns
df_limpio[columnas_texto] = df_limpio[columnas_texto].apply(
    lambda col: col.str.strip().str.upper()
)
```

```python
#Nulos por columna 
nulos = df.isna().sum().sort_values(ascending=False)
display(nulos[nulos > 0])
```

```python
# Porcentaje de nulos: ayuda a decidir después
porcentaje_nulos = (df.isna().mean() * 100).sort_values(ascending=False)
display(porcentaje_nulos[porcentaje_nulos > 0].round(2))
```
